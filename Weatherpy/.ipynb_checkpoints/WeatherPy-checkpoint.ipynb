```python
#import dependencies
import pandas as pd 
import requests
import matplotlib.pyplot as plt
from config import api_key
from config import g_key
import numpy as np
import json
import scipy.stats as st
from scipy.stats import linregress
from citipy import citipy
import time 
import datetime
```

```python
#ranges of latitude and longitude
lat_range=(-90,90)
lng_range=(-180,180)
```

```python
#Create a unique cities list 

lats_lngs=[]
cities=[]

lats=np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs=np.random.uniform(low=-180.000, high=180.000, size=1500)
lats_lngs=zip(lats,lngs)

#For each lat and lng combination, select the closest city 
for lat_lng in lats_lngs:
    city=citipy.nearest_city(lat_lng[0],lat_lng[1]).city_name

    #if the city is unique then we'll add it to our cities data list to create a csv file
    if city not in cities:
        cities.append(city)

len(cities)
```

```python
#insert URL
base_url="http://api.openweathermap.org/data/2.5/weather?"

#creating a counter to go through data 
sets=1
record=0

#Create dataframe
unique_cities=pd.DataFrame({"City":cities})

unique_cities["Cloudiness (%)"]=""
unique_cities["Country"]=""
unique_cities["Date"]=""
unique_cities["Humidity (%)"]=""
unique_cities["Lat"]=""
unique_cities["Lng"]=""
unique_cities["Max Temp (F)"]=""
unique_cities["Wind Speed (mph)"]=""

#go through cities and collect the data 

for index, row in unique_cities.iterrows():
    
    city=row["City"]
    
    query_url=base_url+"appid=" + api_key + "&q=" + city
    cities_response=requests.get(query_url).json()
    
    try:
        print(f"Processing Record {record} of Set {sets} | {city}.") 
        
        unique_cities.loc[index, "Cloudiness (%)"]=cities_response["clouds"]["all"]
        unique_cities.loc[index, "Country"]=cities_response["sys"]["country"]
        unique_cities.loc[index, "Date"]=datetime.datetime.fromtimestamp(cities_response["dt"])
        unique_cities.loc[index, "Humidity (%)"]=cities_response["main"]["humidity"]
        unique_cities.loc[index, "Lat"]=cities_response["coord"]["lat"]
        unique_cities.loc[index, "Lng"]=cities_response["coord"]["lon"]
        unique_cities.loc[index, "Max Temp (F)"]=1.8 * (cities_response["main"]["temp_max"] - 273) + 32
        unique_cities.loc[index, "Wind Speed (mph)"]=cities_response["wind"]["speed"]
        
    except: 
        print(f"City not found .... skipping {city}.")
        
    record=record+1
    if record==50:
        record=0
        sets=sets+1
        
print(f"Data Retrievel Complete")
    
    
```

```python
# Separate cities between the Northern and Southern Hemisphere
northern_hemp=cities_data.loc[(cities_data["Lat"])>0,:]
southern_hemp=cities_data.loc[(cities_data["Lat"])<0,:]
```

```python
#The northern hemp for temp vs lat, with linear regression
x_values=northern_hemp["Lat"]
y_values=northern_hemp["Max Temp (F)"]

(slope, intercept, rvalue, pvalue, stderr)=linregress(x_values,y_values)
regress_values=x_values*slope + intercept 
line_eq="y= " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(f"Line equation is: {line_eq}")

plt.figure(figsize=(10,7))
plt.scatter(x_values,y_values,c="red",edgecolor="black",marker="o")
plt.plot(x_values,regress_values, "r-")
plt.annotate(line_eq,(5,15),color="red",size=15)
plt.xlabel("Latitude",size=15)
plt.ylabel("Temperatures (F)", size=15)
plt.title("Northern Hemisphere: Temperature vs. Latitude", size=20)
plt.savefig('NH_Temp_vs_Lat.png')

print(f"The r-squared value: {rvalue}")
print(f"The further away we move from the equator the temperature decreases.")

plt.show()
```
